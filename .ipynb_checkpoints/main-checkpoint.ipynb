```python
from commonfunctions import *
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2
from skimage.measure import find_contours
from skimage.draw import rectangle
from skimage.morphology import binary_erosion, binary_dilation, binary_closing,skeletonize, thin
```

```python
def flaten_array(array):
    i = 0
    rows = len(array)
    cols = len(array[0])
    flaten_array = np.zeros([1, rows*cols])
    for fx in range(0, rows):
        for fy in range (0, cols):
            flaten_array[0][i] = array[fx][fy]
            i = i + 1
    return np.sort(flaten_array)

def median_filter(window_width, window_height, image):
    edgex = window_width//2
    edgey = window_height//2
    height = len(image)
    width = len(image[0])
    modified_img =  np.copy(image)
    for x in range(edgex, width - edgex):
        for y in range(edgey, height - edgey):
            color_array = np.zeros((window_height, window_width))# the 3x3 window
            for fx in range(0, window_width):
                for fy in range(0, window_height):
                    color_array[fy][fx] = image[y + fy - edgey][x + fx - edgex]
            sorted_array = flaten_array(color_array)
            index = len(sorted_array[0])//2
            modified_img[y][x] = sorted_array[0][index]
    return modified_img
```

```python
imagesFirst=[]

#reading the car image
car=io.imread('images/pic0.jpg')
imagesFirst.append(car)

#getting the dimensions of image
yLength=car.shape[0]
xLength=car.shape[1]

#converting image into gray scale
grayCar=cv2.cvtColor(car,cv2.COLOR_BGR2GRAY)
imagesFirst.append(grayCar)

#applying median for reducing noise
car_denoised=cv2.bilateralFilter(grayCar,11,17,17)
imagesFirst.append(car_denoised)


show_images(imagesFirst)
```

```python
imagesSecond=[]
#using Canny edge detector
edges=canny(car_denoised, sigma=2,low_threshold=40,high_threshold=80)

footPrint=[[0,1,0],[1,1,1],[0,1,0]]

edgesN=binary_dilation(edges,footPrint)

imagesSecond.append(edges)
imagesSecond.append(edgesN)


show_images(imagesSecond)
```

```python
imagesC=[]

contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

print(format(len(contours)))

for c in contours:
    rect=cv2.minAreaRect(c)
    box=cv2.boxPoints(rect)
    box=np.int0(box)
    img=cv2.drawContours(car,[box],0,(0,255,0),3)

imagesC.append(img)

show_images(imagesC)
```

```python


edges=binary_closing(edges)

contours = find_contours(edges,0.1)

bounding_boxes = []
for contour in contours:
    x_cordinats = contour[:,1]
    y_cordinats = contour[:,0]
    
    #---------------------------#
    
    x_max = x_cordinats.max()
    x_min = x_cordinats.min()
    x_difference = x_max - x_min
    y_max = y_cordinats.max()
    y_min = y_cordinats.min()
    y_difference = y_max - y_min
    
    #---------------------------#
    ratio = x_difference / y_difference
    if ratio > 3 and ratio < 8:
        box = [int(np.round(x_min))  , int(np.round(x_max)) , int(np.round(y_min)) , int(np.round(y_max))]
        
        bounding_boxes.append(box)

img_with_boxes = grayCar.copy()

for box in bounding_boxes:
    [Xmin, Xmax, Ymin, Ymax] = box
    rr, cc = rectangle(start = (Ymin,Xmin), end = (Ymax,Xmax), shape=grayCar.shape)
    img_with_boxes[rr, cc] = 1 #set color white

imagess=[]
imagess.append(img_with_boxes)
imagess.append(edges)
show_images(imagess)
```
